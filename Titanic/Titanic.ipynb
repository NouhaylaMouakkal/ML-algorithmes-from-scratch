```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("titanic.csv")
```

```python
data["Sex"] = data["Sex"].map({'male': 0 , 'female': 1})
```

```python
data.head()
```

```python
survived_data = data[data['Survived'] == 1]
not_survived_data = data[data['Survived'] == 0]
```

```python
survived_data
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

sns.histplot(survived_data['Age'], bins=30, kde=True, ax=axes[0])
axes[0].set_title('Survivors per Age)')
axes[0].set_xlabel('Ã‚ge')
axes[0].set_ylabel('FrÃ©quence')

sns.histplot(not_survived_data['Age'], bins=30, kde=True, ax=axes[1])
axes[1].set_title('Unservivors per Age')
axes[1].set_xlabel('Ã‚ge')
axes[1].set_ylabel('FrÃ©quence')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survie en fonction du Sexe')
plt.xlabel('0: Non-survivant, 1: Survivant')
plt.ylabel('Nombre de Passagers')
plt.legend(title='Sexe', labels=['Homme', 'Femme'])
plt.show()
```

```python
sns.pairplot(data, hue='Survived')
plt.show()
```

```python
data["Pclass"].value_counts()
```

```python
plt.figure(figsize=(12, 6))

# Group by 'Pclass' and 'Survived', and calculate counts
survived_by_class = data.groupby(['Pclass', 'Survived']).size().unstack()

# Plot pie chart for each class
for pclass in survived_by_class.index:
    plt.subplot(1, 3, pclass)
    plt.pie(survived_by_class.loc[pclass], labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=90)
    plt.title(f'Class {pclass}')

# Set a layout
plt.tight_layout()

# Show the pie chart
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Fare', data=data)
plt.title('Distribution des Tarifs en fonction de la Survie')
plt.xlabel('Survived (0: Non-survivant, 1: Survivant)')
plt.ylabel('Tarif')
plt.show()
```

```python
class_1_survived = data[(data['Pclass'] == 1) & (data['Survived'] == 1)]
class_1_not_survived = data[(data['Pclass'] == 1) & (data['Survived'] == 0)]

class_2_survived = data[(data['Pclass'] == 2) & (data['Survived'] == 1)]
class_2_not_survived = data[(data['Pclass'] == 2) & (data['Survived'] == 0)]

class_3_survived = data[(data['Pclass'] == 3) & (data['Survived'] == 1)]
class_3_not_survived = data[(data['Pclass'] == 3) & (data['Survived'] == 0)]

# CrÃ©er une figure et un axe
fig, ax = plt.subplots()

# Tracer les lignes pour la classe 1
ax.plot(class_1_survived['Fare'], class_1_survived['Survived'], label='Class 1 - Survived', marker='o', linestyle='-', color='green')
ax.plot(class_1_not_survived['Fare'], class_1_not_survived['Survived'], label='Class 1 - Not Survived', marker='o', linestyle='-', color='red')

# Tracer les lignes pour la classe 2
ax.plot(class_2_survived['Fare'], class_2_survived['Survived'], label='Class 2 - Survived', marker='o', linestyle='-', color='pink')
ax.plot(class_2_not_survived['Fare'], class_2_not_survived['Survived'], label='Class 2 - Not Survived', marker='o', linestyle='-', color='yellow')

# Tracer les lignes pour la classe 3
ax.plot(class_3_survived['Fare'], class_3_survived['Survived'], label='Class 3 - Survived', marker='o', linestyle='-', color='blue')
ax.plot(class_3_not_survived['Fare'], class_3_not_survived['Survived'], label='Class 3 - Not Survived', marker='o', linestyle='-', color='orange')

# Ajouter des lÃ©gendes et des Ã©tiquettes
ax.set_xlabel('Fare')
ax.set_ylabel('Survived')
ax.set_title('Relation entre Fare, Pclass et Survived')
ax.legend()

# Afficher le graphique
plt.show()
```
