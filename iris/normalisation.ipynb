```python
X = dataset[columns].values.reshape(-1, 3)  
Y = dataset['Sales'].values.reshape(-1, 1) 
print(X)
```

```python
product_column1 = np.multiply(X[:, 0], X[:, 1])
product_column2 = np.multiply(X[:, 0], X[:, 2])
product_column3 = np.multiply(X[:, 1], X[:, 2])
product_column4 = np.multiply(X[:, 0], X[:, 0])
product_column5 = np.multiply(X[:, 1], X[:, 1])
product_column6 = np.multiply(X[:, 2], X[:, 2])


X = np.column_stack((product_column1, X))
X = np.column_stack((product_column2, X))
X = np.column_stack((product_column3, X))
X = np.column_stack((product_column4, X))
X = np.column_stack((product_column5, X))
X = np.column_stack((product_column6, X))

print(X.shape)
print(X)
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(X_train)
```

```python
X_train.shape
print(X[:10])
```

```python
#Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
means = scaler.mean_
scales = np.sqrt(scaler.var_)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = (X_test - means)/scales
print(X_test_scaled)
```

```python
ones_column = np.ones((X_train_scaled.shape[0], 1))
ones_column2 = np.ones((X_test_scaled.shape[0], 1))
X_train_scaled = np.hstack((X_train_scaled, ones_column))
X_test_scaled = np.hstack((X_test_scaled, ones_column2))
print(X_test_scaled)
```

```python
np.random.seed(0) # pour produire toujours le meme vecteur theta alÃ©atoire
theta=np.random.randn(10,1)
print('Theta:',theta)
```

```python
def model(X, theta):
    return X.dot(theta)
```

```python
def cost_function(X, y, theta):
    m = len(y)
    return 1/(2*m) * np.sum((model(X, theta) - y)**2)
```

```python
cost_function(X_train_scaled, y_train, theta)
```

```python
def grad(X, y, theta):
    m = len(y)
    return 1/m * X.T.dot(model(X, theta) - y)
```

```python
def gradient_descent(X, y, theta, learning_rate, n_iterations):
    
    cost_history = np.zeros(n_iterations) # crÃ©ation d'un tableau de stockage pour enregistrer l'Ã©volution du Cout du modele
    
    for i in range(0, n_iterations):
        theta = theta - learning_rate * grad(X, y, theta) # mise a jour du parametre theta (formule du gradient descent)
        cost_history[i] = cost_function(X, y, theta) # on enregistre la valeur du Cout au tour i dans cost_history[i]
        
    return theta, cost_history
```

```python
n_iterations = 4000
learning_rate = 0.4

theta_final, cost_history = gradient_descent(X_train_scaled, y_train, theta, learning_rate, n_iterations)
```

```python
theta_final
```

```python
predictions = model(X_train_scaled, theta_final)
predictions
```

```python
plt.plot(range(n_iterations), cost_history)
plt.show()
```

```python
def coef_determination(y, pred):
    u = ((y - pred)**2).sum()
    v = ((y - y.mean())**2).sum()
    return 1 - u/v
```

```python
print(coef_determination(y_train, predictions))
predictions = model(X_test_scaled, theta_final)
print(coef_determination(y_test, predictions))

```
